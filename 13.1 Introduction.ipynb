```python
# Import libraries  Src : GeeksforGeeks
from mpl_toolkits import mplot3d 
import numpy as np 
import matplotlib.pyplot as plt 
  
  
# Creating dataset 
z = np.random.randint(100, size =(50)) 
x = np.random.randint(80, size =(50)) 
y = np.random.randint(60, size =(50)) 
print(str(x[1])+" "+str(y[1])+" "+str(z[1])) 
# Creating figure 
fig = plt.figure(figsize = (10, 7)) 


ax = plt.axes(projection ="3d") 
  
# Creating plot 
ax.scatter3D(x, y, z, color = "green"); 
plt.title("simple 3D scatter plot") 
  
# show plot 
plt.show() 
```

A column vector is an  (n×1) [1]matrix, and a row vector is a  (1×n) matrix


# Data Preprocessing : Column Normalization

The purpose of normalization is, primarily, to scale numeric data from different columns down to an equivalent scale

```python
from sklearn import preprocessing
import numpy as np

a = np.random.random((1, 4))
a = a*20
print("Data = ", a)

# normalize the data attributes
normalized = preprocessing.normalize(a)
print("Normalized Data = ", normalized)
# import matplotlib.pyplot as plt
# plt.plot(normalized)
```

```python
#Simple Normalization Implementation

a = np.arange(1,31)
```

```python
a
```

```python
a_normalized = ((a-np.min(a))/(np.max(a)-np.min(a)))
print(a_normalized)
print(np.min(a))
print(np.max(a))
plt.clf()
plt.plot(a_normalized)
```

```python
#Column Standardization sample
a_standardized = (a-np.mean(a))/(np.std(a))
```
